```python
# 快速傅里叶变换模块(fft)

# 能否用傅里叶分解每天的股价，然后为前十大振幅、前十大频率的正弦函数 赋予事件意义（例如，前十大都比较接近，是什么含义）
# 在音频中，振幅代表音量


import numpy.fft as nf

# 通过采样数与采样周期求得傅里叶变换分解所得曲线的**频率序列**
freqs = np.fft.fftfreq(采样数量, 采样周期)

# 通过原函数值的序列j经过快速傅里叶变换得到一个**复数数组**，复数的模代表的是**振幅**，复数的辐角代表**初相位**
np.fft.fft(原函数值序列) -> 目标函数值序列(复数数组)

# 通过一个复数数组（复数的模代表的是振幅，复数的辐角代表初相位）经过逆向傅里叶变换得到**合成的函数值数组**
np.fft.ifft(目标函数值序列(复数))->原函数值序列
```

```python
# 案例：针对合成波做快速傅里叶变换，得到分解波数组的频率、振幅、初相位数组，并绘制频域图像。

import numpy.fft as nf
import numpy as np
import matplotlib.pyplot as mp


x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.zeros(1000)
n = 1000
for i in range(1, n+1):
    y += 4 / ((2 * i - 1) * np.pi) * np.sin((2 * i - 1) * x)

mp.figure('FFT', facecolor='lightgray')
mp.subplot(121)
mp.title('Time Domain 时域', fontsize=18)
mp.grid(linestyle=':')
mp.plot(x, y, label=r'$y$')
mp.legend()


# 针对方波 y，做 （fft ）变换, 返回（ffts）
complex_ary = nf.fft(y)

# 做逆向变换（ifft）,得到原来的 y
y2 = nf.ifft(complex_ary)   # y2 应该等于 y
mp.plot(x, y2,color='orange', linewidth=5,alpha=0.5, label='y2')

# 得到分解波的频率序列(使用 fft 后complex_ary，绘制频域图像)
mp.subplot(122)
freqs = nf.fftfreq(y.size, x[1] - x[0])  # fftfreq(采样数量, 采样间隔)
# 复数的模（即合向量的长度）为信号的振幅（能量大小）
pows = np.abs(complex_ary)  # abs（ffts）求频率序列数组中每一个向量的模

mp.title('Frequency Domain 频域', fontsize=18)
mp.plot(freqs, pows, color='orangered', label='freqquency')  # freqs[freqs>0] 可以只显示频率为正的部分


mp.legend()
mp.tight_layout()
mp.show()
```

```python
# 基于傅里叶变换的频域滤波

# 案例：基于傅里叶变换的频域滤波为音频文件去除噪声。
# 

import numpy as np
import numpy.fft as nf
import scipy.io.wavfile as wf
import matplotlib.pyplot as mp

# 1. 读取音频文件，获取音频文件基本信息：采样个数，采样周期，与每个采样的声音信号值。绘制音频时域的：时间/位移图像。

sample_rate, noised_sigs = wf.read('/Users/peng/Desktop/ToStudent/month05 datasci/da_data/noised.wav')
noised_sigs = noised_sigs / 2 ** 15   # 范围太大不易观察
times = np.arange(len(noised_sigs)) / sample_rate
mp.figure('Filter', facecolor='lightgray')
mp.subplot(221)
mp.title('Time Domain', fontsize=16)
mp.ylabel('Signal', fontsize=12)
mp.tick_params(labelsize=10)
mp.grid(linestyle=':')
mp.plot(times[:178], noised_sigs[:178],c='orangered', label='Noised')   # 取前 178 个采样点绘图】】】】】】】】】】】】】】】】】】】】】】】】】
mp.legend()
mp.show()


# 2. 基于傅里叶变换，获取音频频域信息，绘制音频频域的：频率/能量图像。

freqs = nf.fftfreq(times.size, 1 / sample_rate)
noised_ffts = nf.fft(noised_sigs)
noised_pows = np.abs(noised_ffts)
mp.subplot(222)
mp.title('Frequency Domain', fontsize=16)
mp.ylabel('Power', fontsize=12)
mp.tick_params(labelsize=10)
mp.grid(linestyle=':')
mp.semilogy(freqs[freqs >= 0],noised_pows[freqs >= 0], c='limegreen',label='Noised')
mp.legend()


# 3. 将低能噪声去除后绘制音频频域的：频率/能量图像。

fund_freq = freqs[noised_pows.argmax()]
noised_indices = np.where(freqs != fund_freq)    # 获取低能音频的下标
filter_ffts = noised_ffts.copy()
filter_ffts[noised_indices] = 0
filter_pows = np.abs(filter_ffts)

mp.subplot(224)
mp.xlabel('Frequency', fontsize=12)
mp.ylabel('Power', fontsize=12)
mp.tick_params(labelsize=10)
mp.grid(linestyle=':')
mp.plot(freqs[freqs >= 0], filter_pows[freqs >= 0],c='dodgerblue', label='Filter')
mp.legend() 


# 4. 基于逆向傅里叶变换，生成新的音频信号，绘制音频时域的：时间/位移图像。

filter_sigs = nf.ifft(filter_ffts).real
mp.subplot(223)
mp.xlabel('Time', fontsize=12)
mp.ylabel('Signal', fontsize=12)
mp.tick_params(labelsize=10)
mp.grid(linestyle=':')
mp.plot(times[:178], filter_sigs[:178],c='hotpink', label='Filter')
mp.legend()


# 5. 重新生成音频文件。

wf.write('/Users/peng/Desktop/ToStudent/month05 datasci/da_data/filter.wav',
         sample_rate,
         (filter_sigs * 2 ** 15).astype(np.int16))
```

```python
# 随机数模块(random)

# 二项分布（binomial）   (每次发生的概率独立，且，每次都只有两种情况)
import numpy as np

# 1. 某人投篮命中率为0.3，投10次，进5个球的概率。
n = np.random.binomial(10, 0.3, 1)   # 参数分别为实验次数， 定义事件发生的概率， 重复多少轮完整的实验
print(n)  

p_list = []
n = np.random.binomial(10, 0.3, 10000)   # 参数分别为样本空间， 定义事件发生的概率， 重复多少轮完整的实验（产生的输出数）
for i in range(0, 11):
    p = (n==i).sum()/10000    # 可以发现，随着实验次数的增加，均值会趋于稳定  （n=3 的概率会最高）
    print("发生",i,"次的概率为：",p)     # 总是在 0.1 附近
    p_list.append(p)

x = range(0, 11)
mp.plot(x, p_list, linestyle=':')
```

```python
# 超几何分布(hypergeometric)   （无放回的抽样）

# 每个随机数t为在总样本中随机抽取nsample个样本后好样本的个数，总样本由ngood个好样本和nbad个坏样本组成，产生size个随机数
# np.random.hypergeometric(ngood, nbad, nsample, size)


p_list = []
n = np.random.hypergeometric(7, 3, 5, 20000)   # 参数分别为正面样本数，反面样本数，正面事件，重复多少轮完整的实验（产生的输出数）
for i in range(2, 6):    # 只有 2、3、4、5 其中一个可能
    p = (n==i).sum()/20000    # 可以发现，随着实验次数的增加，会趋近正态分布  
    print("发生",i,"次的概率为：",p)     # 总是在 哪里？ 附近
    p_list.append(p)

x = range(2,  6)
mp.plot(x, p_list, linestyle=':')
```
