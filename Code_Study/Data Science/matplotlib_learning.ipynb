```python
locators = ['mp.NullLocator()', 
            'mp.MaxNLocator(nbins=4)',
            'mp.FixedLocator(locs=[0, 2.5, 5, 7.5, 10])',
            'mp.AutoLocator()',
            'mp.IndexLocator(offset=0.5, base=1.5)',
            'mp.MultipleLocator()',
            'mp.LinearLocator(numticks=21)',
            'mp.LogLocator(base=2)'
           ]
mp.figure('Locator', facecolor='gray')


for i, locator in enumerate(locators):   # 枚举遍历，不仅返回对象，还返回一个索引值，
    mp.subplot(len(locators), 1, i+1)    # 画len(locators)个子图，每个占一行
    mp.xlim(0, 10)
    mp.ylim(-1, 1)
    mp.yticks([])
    
    ax = mp.gca()
    ax.spines['left'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    # 将底坐标轴调整到子图中心位置
    ax.spines['bottom'].set_position(('data', 0))
    # 设置水平坐标轴的主刻度定位器
    ax.xaxis.set_major_locator(eval(locator))
    # 设置水平坐标轴的次刻度定位器为多点定位器，间隔0.1
    ax.xaxis.set_minor_locator(mp.MultipleLocator(0.1))
    mp.plot(np.arange(11), np.zeros(11), c='none')
    # 标记所用刻度定位器类名
    mp.text(0, 0.3, locator, ha='left', size=12)
    
    mp.tight_layout()
    mp.show()
    
```

```python
# 填充:以某种颜色自动填充两条曲线的闭合区域。fill_between()

n = 1000
x = np.linspace(0, 8 * np.pi, n)
sin_y = np.sin(x)
cos_y = np.cos(x / 2) / 2

mp.figure('Fill', facecolor='lightgray')
mp.title('Fill', fontsize=20)
mp.xlabel('x', fontsize=14)
mp.ylabel('y', fontsize=14)

mp.tick_params(labelsize=10)
mp.grid(linestyle=':')

mp.plot(x, sin_y, c='dodgerblue',
        label=r'$y=sin(x)$')
mp.plot(x, cos_y, c='orangered',
        label=r'$y=\frac{1}{2}cos(\frac{x}{2})$')

# 填充
# mp.fill_between(
#     x,				# x轴的水平坐标
#     sin_x,			# 下边界曲线上点的垂直坐标
#     cos_x,			# 上边界曲线上点的垂直坐标
#     sin_x<cos_x, 	    # 填充条件，为True时填充
#     color='', 		# 填充颜色
#     alpha=0.2		    # 透明度
# )
mp.fill_between(x, cos_y, sin_y, cos_y < sin_y,
                color='dodgerblue', alpha=0.5)
mp.fill_between(x, cos_y, sin_y, cos_y > sin_y,
                color='orangered', alpha=0.5)

mp.legend()
mp.show()
```

```python
# 等高线图

# mp.contourf(x, y, z, 8, cmap='jet')
# cntr = mp.contour(
#     x, 					# 网格坐标矩阵的x坐标 （2维数组）
#     y, 					# 网格坐标矩阵的y坐标 （2维数组）
#     z, 					# 网格坐标矩阵的z坐标 （2维数组）
#     8, 					# 把等高线绘制成8部分
#     colors='black',		# 等高线的颜色
# 	linewidths=0.5		# 线宽
# )


n = 1000

# 生成网格化坐标矩阵(分布生成 1000 个 x、y，再生成（x，y）坐标)
x, y = np.meshgrid(np.linspace(-3, 3, n),
                   np.linspace(-3, 3, n))

# 根据每个网格点坐标，通过某个公式计算z高度坐标
z = (1 - x/2 + x**5 + y**3) * np.exp(-x**2 - y**2)

mp.figure('Contour', facecolor='lightgray')
mp.title('Contour', fontsize=20)
mp.xlabel('x', fontsize=14)
mp.ylabel('y', fontsize=14)
mp.tick_params(labelsize=10)
mp.grid(linestyle=':')

# 绘制等高线图
mp.contourf(x, y, z, 8, cmap='jet')   # ’8‘ 为等高线的数量
cntr = mp.contour(x, y, z, 8, colors='black',
                  linewidths=0.5)

# 为等高线图添加高度标签（高度等）
mp.clabel(cntr, inline_spacing=1, fmt='%.1f',
          fontsize=10)

mp.show()
```

```python
# 热成像图（方便直观阅读大的数据集）

# 生成网格化坐标矩阵(分布生成 1000 个 x、y，再生成（x，y）坐标)
x, y = np.meshgrid(np.linspace(-3, 3, n),
                   np.linspace(-3, 3, n))

# 根据每个网格点坐标，通过某个公式计算z高度坐标
z = (1 - x/2 + x**5 + y**3) * np.exp(-x**2 - y**2)

mp.title('Imshow',fontsize=15)

# imshow（）把矩阵z图形化，使用cmap表示矩阵中每个元素值的大小
# origin: 坐标轴方向
#    upper: 缺省值，原点在左上角
#    lower: 原点在左下角
mp.imshow(z, cmap='jet', origin='low')

# 显示辅助的颜色条提示
mp.colorbar()
mp.show()
```

```python
# 简单动画（映射实时数据）

import matplotlib.animation as ma

#定义更新函数行为
# def update(number):
#     pass
# 每隔10毫秒执行一次update更新函数，作用于mp.gcf()当前窗口对象
# mp.gcf()：	获取当前窗口
# update：	更新函数
# interval：	间隔时间（单位：毫秒）
# anim = ma.FuncAnimation(mp.gcf(), update, interval=10)
# mp.show()


# 案例：随机生成各种颜色的100个气泡。让他们不断的增大。

#自定义一种可以存放在ndarray里的类型class，用于保存一个气泡
ball_type = np.dtype([
	('position', float, 2),  # 位置(水平和垂直坐标)
    ('size', float, 1),      # 大小
    ('growth', float, 1),    # 生长速度
    ('color', float, 4)])    # 颜色(红、绿、蓝和透明度)

#随机生成100个点对象
n = 100
balls = np.zeros(100, dtype=ball_type)
balls['position']=np.random.uniform(0, 1, (n, 2))  # 生成(n * 2)的数组，服从均匀分布
balls['size']=np.random.uniform(40, 70, n)
balls['growth']=np.random.uniform(10, 20, n)
balls['color']=np.random.uniform(0, 1, (n, 4))

# 绘制图像
mp.figure("Animation", facecolor='lightgray')
mp.title("Animation", fontsize=14)
mp.xticks([])
mp.yticks([])

sc = mp.scatter(
	balls['position'][:, 0],   # [:, 0]所有行的第一列
	balls['position'][:, 1],   # [:, 1]所有行的第二列
	balls['size'], 
	color=balls['color'], alpha=0.5)

#定义更新函数行为（动画）
def update(number):
    # 更新每个泡泡的 size
	balls['size'] += balls['growth']
	#每次让一个气泡破裂，随机生成一个新的
	boom_ind = number % n
	balls[boom_ind]['size']=np.random.uniform(40, 70, 1)
	balls[boom_ind]['position']=np.random.uniform(0, 1, (1, 2))
	# 重新设置属性
	sc.set_sizes(balls['size'])
	sc.set_offsets(balls['position'])

# 每隔30毫秒执行一次update更新函数，作用于mp.gcf()当前窗口图片对象（动画的本质）
# mp.gcf()：	获取当前窗口
# update：		更新函数
# interval：	间隔时间（单位：毫秒）
anim = ma.FuncAnimation(mp.gcf(), update, interval=70)
mp.show()
```

```python
# 标准差


# 样本：S = [s1, s2, ..., sn]
# 平均值：m = (s1+s2+...+sn)/n
# 离差：D = [d1, d2, ..., dn], di = si-m  （离开均值的正负偏差累计，可抵消）
# 离差方：Q = [q1, q2, ..., qn], qi = di2
# 总体方差：v = (q1+q2+...+qn)/n      （离开均值的偏差累计）
# 总体标准差：s = sqrt(v)，方均根
# 样本方差：v' = (q1+q2+...+qn)/(n-1)   （比总体方差大）
# 样本标准差：s' = sqrt(v')，方均根

import numpy as np
closing_prices = np.loadtxt(
    '/Users/peng/Desktop/ToStudent/month05 datasci/da_data/aapl.csv', 
    delimiter=',', usecols=(6), unpack=True)

mean = np.mean(closing_prices)         # 算数平均值

devs = closing_prices - mean           # 离差

dsqs = devs ** 2                       # 离差方

pvar = np.sum(dsqs) / dsqs.size        # 总体方差

pstd = np.sqrt(pvar)                   # 总体标准差

svar = np.sum(dsqs) / (dsqs.size - 1)  # 样本方差
sstd = np.sqrt(svar)                   # 样本标准差
print("svar:",svar, "sstd:",sstd)

pstd = np.std(closing_prices)          # 总体标准差
sstd = np.std(closing_prices, ddof=1)  # 样本标准差
print("pstd:",pstd, "sstd:",sstd)
```

```python
# 线性模型


# 什么是线性关系？  （1、2、3、4 维）直线、平面、立方体、超立方体

# np.linalg.lstsq(A, B)   # lstsq 是 LeaST SQuare （最小二乘）的意思（返回一个元组、求权重向量：未知）
import pandas as pd
import numpy as np

# 整理五元一次方程组    最终获取一组股票走势预测值

N = 5
pred_prices = np.zeros(closing_prices.size - 2 * N + 1)  #  最后一个，多一天
for i in range(pred_prices.size):
    a = np.zeros((N, N))
    for j in range(N):
        a[j, ] = closing_prices[i + j:i + j + N]
    b = closing_prices[i + N:i + N * 2]
    x = np.linalg.lstsq(a, b)[0]
    pred_prices[i] = b.dot(x)
    
# 由于预测的是下一天的收盘价，所以想日期数组中追加一个元素，为下一个工作日的日期
dates = dates.astype(md.datetime.datetime)
mp.plot(dates, closing_prices, 'o-', c='lightgray', label='Closing Price')
dates = np.append(dates, dates[-1] + pd.tseries.offsets.BDay())
mp.plot(dates[2 * N:], pred_prices, 'o-',c='orangered', 
        linewidth=3,label='Predicted Price')


mp.legend()
mp.gcf().autofmt_xdate() 
mp.show()
```

```python
# 协方差、相关矩阵、相关系数

# 评估这两组统计数据的相似程度。
# 协方差  可以用离差计算（两组数据同时偏离各自均值吗）
# 相关系数  协方差的归一化（通过标准差的积作为分母归一化（若分子为标准差的积，则相关系数为一））

# 协方差可以简单反映两组统计样本的相关性，
# 值为正，则为正相关；值为负，则为负相关，绝对值越大相关性越强。越接近 0 则两组数据越独立

# cov_ab = ave(dev_a * dev_b)  （两组数据同时偏离各自均值吗）
# cov_ba = ave(dev_b * dev_a)  （两组数据同时偏离各自均值吗）


import numpy as np
import matplotlib.pyplot as mp
import datetime as dt
import matplotlib.dates as md

def dmy2ymd(dmy):
    dmy = str(dmy, encoding='utf-8')
    t = dt.datetime.strptime(dmy, '%d-%m-%Y')
    s = t.date().strftime('%Y-%m-%d')
    return s

dates, bhp_closing_prices = \
np.loadtxt('/Users/peng/Desktop/ToStudent/month05 datasci/da_data/bhp.csv',
           delimiter=',',
           usecols=(1,6),
           unpack=True,
           dtype='M8[D], f8',
           converters={1:dmy2ymd}   # 像 map 函数, 预处理第一列
          )

vale_closing_prices = \
np.loadtxt('/Users/peng/Desktop/ToStudent/month05 datasci/da_data/vale.csv',
           delimiter=',',
           usecols=(6,),
           unpack=True,
           dtype='f8',
#            converters={1:dmy2ymd}   # 像 map 函数, 预处理第一列
          )

ax = mp.gca()
ax.xaxis.set_major_locator(md.WeekdayLocator(byweekday=md.MO))
ax.xaxis.set_major_formatter(md.DateFormatter('%Y-%m-%d'))
ax.xaxis.set_minor_locator(md.DayLocator())

mp.plot(dates, bhp_closing_prices, color='dodgerblue', linewidth=0.7, alpha=0.8, label='BHP CPrice')
mp.plot(dates, vale_closing_prices, color='blue', linewidth=0.7, alpha=0.8, label='VALE CPrice')


# 计算两只股票的协方差 cov
mean_bhp = np.mean(bhp_closing_prices)
mean_vale = np.mean(vale_closing_prices)
dev_bhp = bhp_closing_prices - mean_bhp
dev_vale = vale_closing_prices - mean_vale
cov = (dev_bhp * dev_vale).mean()
print('cov:', cov)

mp.legend()
mp.gcf().autofmt_xdate() # 自适应日期显示？
mp.show()
```

```python
# 相关矩阵   np.corrcoef(A,B)
# 使用相关矩阵获得两组数据的相关系数(协方差矩阵刚好是相关矩阵的分子)   矩阵正对角线上的值都为1。（同组样本自己相比绝对正相关）

k = np.corrcoef(bhp_closing_prices,vale_closing_prices)
cov_matrix = np.cov(bhp_closing_prices,vale_closing_prices)   # 相关矩阵的分子矩阵 （作为相关矩阵的分子）
print(k,'\n', cov_matrix)
# [[a方差，ab协方差], 
#  [ba协方差, b方差]]  协方差比直接计算出来的大（总体的小）
```

```python
# 多项式拟合

# 多项式拟合的目的是为了找到一组p0-pn，使得拟合方程尽可能的与实际样本数据相符合
############ API
# X = [x1, x2, ..., xn] - 自变量
# Y = [y1, y2, ..., yn] - 实际函数值
# Y'= [y1',y2',...,yn'] - 拟合函数值
# P = [p0, p1, ..., pn] - 多项式函数中的系数

# 根据一组样本，并给出最高次幂，求出拟合系数P（数组）
# np.polyfit(X, Y, 最高次幂)->P

# 根据拟合系数P 与任意自变量X 求出拟合值Y', 由此可得拟合曲线坐标样本数据 [X, Y']
# np.polyval(P, X)->Y'

# 多项式函数求导，根据拟合系数求出多项式函数导函数的系数数组
# np.polyder(P)->Q 

# 已知多项式系数Q 求多项式函数的根（与x轴交点的横坐标）---导函数的零点
# xs = np.roots(Q)

# 两个多项式函数的差函数的系数（可以通过差函数的根求取两个曲线的交点） （差函数等于 0）即两函数 y 相等的点，就是求差函数的根
# Q = np.polysub(P1, P2)



import numpy as np
import matplotlib.pyplot as mp

x = np.linspace(-20, 20, 1000)  
y = 4*x**3 + 3*x**2  - 1000*x + 1
Q = np.polyder([4,3,-1000,1])
xs = np.roots(Q)
ys =  4*xs**3 + 3*xs**2  - 1000*xs + 1
mp.plot(x, y)
mp.scatter(xs, ys, s=80, c='orangered')
mp.show()
```
